```python
import numpy as np
from matplotlib import pyplot

fig, ax = pyplot.subplots(constrained_layout=True, figsize=(10, 10))

xi = np.arange(100000) + 1
box_width = np.linspace(1.0, 10.0, 1000)

sinc_sum = box_width * (np.sum(np.sinc(xi[np.newaxis, :] * box_width[:, np.newaxis]), axis=-1) * 2 + 1)

path = ax.scatter(box_width, sinc_sum, s=1)
```
