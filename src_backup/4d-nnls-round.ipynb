All the possible images $R(x_r, y_r, z_f)$ formed on the retina for different focus distances $z_f$ can be defined from the light field $L(x_p, y_p, u, v)$ over the pupil:
$$ \begin{align*}
    R(x_r, y_r, z_f) = \iint_{-\infty}^{\infty} L\!\left( x_p, y_p, \frac{x_r}{z_r} - \frac{x_p}{z_f}, \frac{y_r}{z_r} - \frac{y_p}{z_f} \right) \, O\!\left( x_r, y_r, x_p, y_p \right) \, \mathrm{d}x_p \, \mathrm{d}y_p && (1)\\
\end{align*} $$

Where $O(x_r, y_r, x_p, y_p)$ is responsible for the finite pupil and angle-dependent effects such as vignetting.

The light field $\tilde{L}(x_p, y_p, u, v)$ over the pupil that was emitted by the display can be defined as:
$$ \begin{align*}
    \tilde{L}(x_p, y_p, u, v) = \sum_{i = 1}^{n} c_{i} \, G_i\!\left( x_p, y_p, u, v \right) && (2)\\
\end{align*} $$

Where each display element $i$ emits a constant light field $G_i(x_p, y_p, u, v)$ that is attenuated by a coefficient $c_i \in [0, 1]$.

The resulting retina images $\tilde{R}(x_r, y_r, z_f)$ from observing such display can be defined as:
$$ \begin{align*}
    \tilde{R}(x_r, y_r, z_f) &= \iint_{-\infty}^{\infty} \tilde{L}\!\left( x_p, y_p, \frac{x_r}{z_r} - \frac{x_p}{z_f}, \frac{y_r}{z_r} - \frac{y_p}{z_f} \right) \,
                                                         O\!\left( x_r, y_r, x_p, y_p \right) \, \mathrm{d}x_p \, \mathrm{d}y_p && \therefore\\
    \tilde{R}(x_r, y_r, z_f) &= \iint_{-\infty}^{\infty} \left( \sum_{i = 1}^{n} c_{i} \, G_i\!\left( x_p, y_p, \frac{x_r}{z_r} - \frac{x_p}{z_f}, \frac{y_r}{z_r} - \frac{y_p}{z_f} \right) \right) \,
                                                         O\!\left( x_r, y_r, x_p, y_p \right) \, \mathrm{d}x_p \, \mathrm{d}y_p && \therefore\\
    \tilde{R}(x_r, y_r, z_f) &= \sum_{i = 1}^{n} c_{i} \, \iint_{-\infty}^{\infty} G_i\!\left( x_p, y_p, \frac{x_r}{z_r} - \frac{x_p}{z_f}, \frac{y_r}{z_r} - \frac{y_p}{z_f} \right) \,
                                                                                   O\!\left( x_r, y_r, x_p, y_p \right) \, \mathrm{d}x_p \, \mathrm{d}y_p && \therefore\\
    \tilde{R}(x_r, y_r, z_f) &= \sum_{i = 1}^{n} c_{i} \, \varphi_i\!\left( x_r, y_r, z_f \right) = \Phi\boldsymbol{c} && (3)\\
\end{align*} $$

Where:
$$ \begin{align*}
    \varphi_i(x_r, y_r, z_f) &= \iint_{-\infty}^{\infty} G_i\!\left( x_p, y_p, \frac{x_r}{z_r} - \frac{x_p}{z_f}, \frac{y_r}{z_r} - \frac{y_p}{z_f} \right) \,
                                                         O\!\left( x_r, y_r, x_p, y_p \right) \, \mathrm{d}x_p \, \mathrm{d}y_p && (4)\\
\end{align*} $$

For a given desired retina image $R(x_r, y_r, z_f)$, we can formulate the following optimization problem:
$$ \begin{align*}
    & \underset{\boldsymbol{c}}{\text{minimize}} && \big\lVert R - \Phi \boldsymbol{c} \big\rVert_2 && (5)\\
    & \text{subject to} && 0 \leq c_i \leq 1, \quad i = 1, \ldots, n
\end{align*} $$

To solve (5) we employ an iterative multiplicative rule:
$$ \begin{align*}
    \boldsymbol{c} \leftarrow \boldsymbol{c} \, \circ \, \frac{\Phi^\intercal R}{\Phi^\intercal \Phi \boldsymbol{c}} && (6)\\
\end{align*} $$

Where:
$$ \begin{align*}
    \left[ \Phi^\intercal R \right]_{i} = \langle \varphi_{i}, R \rangle &=
    \iiint_{-\infty}^{\infty} \varphi_i\!\left( x_r, y_r, z_f \right) \, R(x_r, y_r, z_f) \, \mathrm{d}x_r \, \mathrm{d}y_r \, \mathrm{d}z_f && (7)\\
    \left[ \Phi^\intercal \Phi \right]_{i, j} = \langle \varphi_{i}, \varphi_{j} \rangle &=
    \iiint_{-\infty}^{\infty} \varphi_i\!\left( x_r, y_r, z_f \right) \, \varphi_j\!\left( x_r, y_r, z_f \right) \, \mathrm{d}x_r \, \mathrm{d}y_r \, \mathrm{d}z_f && (8)\\
\end{align*} $$

Substituting (1) in (7):
$$ \begin{align*}
    \left[ \Phi^\intercal R \right]_{i} &= \iiint_{-\infty}^{\infty} \varphi_{i}\!\left( x_r, y_r, z_f \right) \,
        \left( \iint_{-\infty}^{\infty} L\!\left( x_p, y_p, \frac{x_r}{z_r} - \frac{x_p}{z_f}, \frac{y_r}{z_r} - \frac{y_p}{z_f} \right) \, O\!\left( x_r, y_r, x_p, y_p \right) \, \mathrm{d}x_p \, \mathrm{d}y_p \right) \,
                                                                     \mathrm{d}x_r \, \mathrm{d}y_r \, \mathrm{d}z_f && \therefore \\
    \left[ \Phi^\intercal R \right]_{i} &= \iint\!\!\!\iiint_{-\infty}^{\infty} \varphi_{i}\!\left( \frac{z_r}{z_f} x_p + z_r u, \frac{z_r}{z_f} y_p + z_r v, z_f \right) \, L\!\left( x_p, y_p, u, v \right) \,
                                                                                O\!\left( \frac{z_r}{z_f} x_p + z_r u, \frac{z_r}{z_f} y_p + z_r v, x_p, y_p \right) \,
                                                                                z_r^2 \, \mathrm{d}x_p \, \mathrm{d}y_p \, \mathrm{d}u \, \mathrm{d}v \, \mathrm{d}z_f && \therefore \\
    \left[ \Phi^\intercal R \right]_{i} &= \iint\!\!\!\iint_{-\infty}^{\infty} L(x_p, y_p, u, v) \left( z_r^2 \int_{-\infty}^{\infty} \varphi_{i}\!\left( \frac{z_r}{z_f} x_p + z_r u, \frac{z_r}{z_f} y_p + z_r v, z_f \right) \,
                                                                                                        O\!\left( \frac{z_r}{z_f} x_p + z_r u, \frac{z_r}{z_f} y_p + z_r v, x_p, y_p \right) \, \mathrm{d}z_f \right) \,
                                                                               \mathrm{d}x_p \, \mathrm{d}y_p \, \mathrm{d}u \, \mathrm{d}v && \therefore \\
    \left[ \Phi^\intercal R \right]_{i} &= \iint\!\!\!\iint_{-\infty}^{\infty} L(x_p, y_p, u, v) \, \eta_i(x_p, y_p, u, v) \, \mathrm{d}x_p \, \mathrm{d}y_p \, \mathrm{d}u \, \mathrm{d}v && (9)\\
\end{align*} $$

Where:
$$ \begin{align*}
    \eta_i(x_p, y_p, u, v) &= z_r^2 \int_{-\infty}^{\infty} \varphi_{i}\!\left( \frac{z_r}{z_f} x_p + z_r u, \frac{z_r}{z_f} y_p + z_r v, z_f \right) \,
                                                            O\!\left( \frac{z_r}{z_f} x_p + z_r u, \frac{z_r}{z_f} y_p + z_r v, x_p, y_p \right) \, \mathrm{d}z_f && (10)\\
\end{align*} $$

Therefore, $\eta_i(x_p, y_p, u, v)$ measures how much influence a given ray of the light field $L(x_p, y_p, u, v)$ has over the i-th display element.

---

If $\varphi$ is separable, we have:

$$ \begin{align*}
    \varphi_i\!\left( x_r, y_r, z_f \right) = \varphi^x_i\!\left( x_r, z_f \right) \, \varphi^y_i\!\left( y_r, z_f \right) && (11)\\
\end{align*} $$

Therefore:

$$ \begin{align*}
    \left[ \Phi^\intercal \Phi \right]_{i, j} &= \iiint_{-\infty}^{\infty} \varphi_i\!\left( x_r, y_r, z_f \right) \, \varphi_j\!\left( x_r, y_r, z_f \right) \, \mathrm{d}x_r \, \mathrm{d}y_r \, \mathrm{d}z_f && \therefore\\
    \left[ \Phi^\intercal \Phi \right]_{i, j} &= \int_{-\infty}^{\infty} \left( \int_{-\infty}^{\infty} \varphi^x_i\!\left( x_r, z_f \right) \, \varphi^x_j\!\left( x_r, z_f \right) \, \mathrm{d}x_r \right)
                                                                         \left( \int_{-\infty}^{\infty} \varphi^y_i\!\left( y_r, z_f \right) \, \varphi^y_j\!\left( y_r, z_f \right) \, \mathrm{d}y_r \right) \mathrm{d}z_f && (12)\\
\end{align*} $$

The one-dimensional $\varphi^x$ can be defined as:

$$ \begin{align*}
    \varphi^x_i\!\left( x_r, z_f \right) = \sum_{k = 1}^{4} \gamma^x_{i,k}\!\left( x_r, z_f \right) && (13)\\
\end{align*} $$

Where each $\gamma$ function corresponds to one of the edges of the one-dimensional display element.

From (13) we have:

$$ \begin{align*}
    \int_{-\infty}^{\infty} \varphi^x_i\!\left( x_r, z_f \right) \, \varphi^x_j\!\left( x_r, z_f \right) \, \mathrm{d}x_r &=
    \int_{-\infty}^{\infty} \left( \sum_{k = 1}^{4} \gamma^x_{i,k}\!\left( x_r, z_f \right) \right) \left( \sum_{l = 1}^{4} \gamma^x_{j,l}\!\left( x_r, z_f \right) \right) \mathrm{d}x_r && \therefore\\
    \int_{-\infty}^{\infty} \varphi^x_i\!\left( x_r, z_f \right) \, \varphi^x_j\!\left( x_r, z_f \right) \, \mathrm{d}x_r &=
    \sum_{k = 1}^{4} \sum_{l = 1}^{4} \int_{-\infty}^{\infty} \gamma^x_{i,k}\!\left( x_r, z_f \right) \, \gamma^x_{j,l}\!\left( x_r, z_f \right) \, \mathrm{d}x_r && (14)\\
\end{align*} $$

The position $r_{\underline{a_i}, \underline{b_i}}\!\left( z_f \right)$ at the retina that corresponds to the corner $(\underline{a_i}, \underline{b_i})$ of the i-th one-dimensional display element is given by:

$$ \begin{align*}
    r_{\underline{a_i}, \underline{b_i}}\!\left( z_f \right) &= \frac{-z_r}{z_b - z_a} \left[ \left( 1 - \frac{z_b}{z_f} \right) \underline{a_i} - \left( 1 - \frac{z_a}{z_f} \right) \underline{b_i} \right] && \therefore\\
    r_{\underline{a_i}, \underline{b_i}}\!\left( z_f \right) &= \frac{-z_r}{z_b - z_a} \left[ \left( \underline{a_i} - \underline{b_i} \right) + \left(z_a \underline{b_i} - z_b \underline{a_i} \right) z^{-1}_f \right] && (15)\\
\end{align*} $$

The positions $r_{\underline{a_i}, \overline{b_i}}\!\left( z_f \right)$, $r_{\overline{a_i}, \underline{b_i}}\!\left( z_f \right)$ and $r_{\overline{a_i}, \overline{b_i}}\!\left( z_f \right)$ are defined analogously.

Then, the $\gamma$ function associated with the edge $x_a = \underline{a_i}$, denoted $\gamma^x_{\underline{a_i}}\!\left( x_r, z_f \right)$, is given by:

$$ \begin{align*}
    \gamma^x_{\underline{a_i}}\!\left( x_r, z_f \right) = \frac{1}{1 - z_a z^{-1}_f} \left( \frac{z_a}{-z_r} x_r + \underline{a_i} \right)
    \quad \text{if} \quad x_r \in \left[ r_{\underline{a_i}, \overline{b_i}}\!\left( z_f \right), r_{\underline{a_i}, \underline{b_i}}\!\left( z_f \right) \right]
    \quad , \quad 0 \quad \text{otherwise} && (16)\\
\end{align*} $$

The functions $\gamma^x_{\overline{a_i}}\!\left( x_r, z_f \right)$, $\gamma^x_{\underline{b_i}}\!\left( x_r, z_f \right)$ and $\gamma^x_{\overline{b_i}}\!\left( x_r, z_f \right)$ are defined analogously.

The limits of integration in (14) will depend on $z_f$ since the support of $\gamma^x_{i,k}\!\left( x_r, z_f \right)$ and $\gamma^x_{j,l}\!\left( x_r, z_f \right)$ depend on $r_{\boldsymbol{\cdot}, \boldsymbol{\cdot}}\!\left( z_f \right)$.
Due to its discontinuities, this integral must be split into at most 4 pieces, each with corresponding integration limits of $\underline{r}\left( z_f \right)$ and $\overline{r}\left( z_f \right)$.

One of such pieces for the term with $\gamma^x_{\underline{a_i}}\!\left( x_r, z_f \right)$ and $\gamma^x_{\underline{b_j}}\!\left( x_r, z_f \right)$, for example, is given by:

$$ \begin{align*}
    \int_{\underline{r}\left( z_f \right)}^{\overline{r}\left( z_f \right)} \gamma^x_{\underline{a_i}}\!\left( x_r, z_f \right) \, \gamma^x_{\underline{b_j}}\!\left( x_r, z_f \right) \, \mathrm{d}x_r &=
    \frac{1}{1 - z_a z^{-1}_f} \frac{1}{1 - z_b z^{-1}_f} \int_{\underline{r}\left( z_f \right)}^{\overline{r}\left( z_f \right)} \left[ \frac{z_a}{-z_r} \frac{z_b}{-z_r} x^2_r +
                                                                                                                               \left( \frac{z_b}{-z_r} \underline{a_i} + \frac{z_a}{-z_r} \underline{b_j} \right) x_r +
                                                                                                                               \underline{a_i} \underline{b_j} \right] \mathrm{d}x_r && \therefore\\
    \int_{\underline{r}\left( z_f \right)}^{\overline{r}\left( z_f \right)} \gamma^x_{\underline{a_i}}\!\left( x_r, z_f \right) \, \gamma^x_{\underline{b_j}}\!\left( x_r, z_f \right) \, \mathrm{d}x_r &=
    \frac{1}{1 - z_a z^{-1}_f} \frac{1}{1 - z_b z^{-1}_f} \left[ \frac{z_a}{-z_r} \frac{z_b}{-z_r} \frac{x^3_r}{3} +
                                                       \left( \frac{z_b}{-z_r} \underline{a_i} + \frac{z_a}{-z_r} \underline{b_j} \right) \frac{x^2_r}{2} +
                                                       \underline{a_i} \underline{b_j} x_r \bigg|_{\underline{r}\left( z_f \right)}^{\overline{r}\left( z_f \right)} \right] && (17)\\
\end{align*} $$

Since $r_{\boldsymbol{\cdot}, \boldsymbol{\cdot}}\!\left( z_f \right)$ is hyperbolic in $z_f$, Equation (17) is a rational function over $z^{-1}_f$ with a polynomial numerator of degree 3 and a polynomial denominator of degree 2.

Therefore, the integral in Equation (12) is of the form:

$$ \begin{align*}
    \left[ \Phi^\intercal \Phi \right]_{i, j} = \sum_{1}^{4^6} \int_{\underline{z_f}}^{\overline{z_f}} \frac{p_6 z^{-6}_f + p_5 z^{-5}_f + p_4 z^{-4}_f + p_3 z^{-3}_f + p_2 z^{-2}_f + p_1 z^{-1}_f + p_0}
                                                                                                            {q_4 z^{-4}_f + q_3 z^{-3}_f + q_2 z^{-2}_f + q_1 z^{-1}_f + q_0} \, \mathrm{d}z_f && (18)\\
\end{align*} $$

---

Space transformations:

$$ \begin{align*}
    \begin{bmatrix} x_a \\ x_b \end{bmatrix} &= \begin{bmatrix} \frac{z_a}{z_r} & 1 - \frac{z_a}{z_f} \\
                                                                \frac{z_b}{z_r} & 1 - \frac{z_b}{z_f} \end{bmatrix} \begin{bmatrix} x_r \\ x_p \end{bmatrix} \\
    \begin{bmatrix} x_r \\ x_p \end{bmatrix} &= \begin{bmatrix} -\frac{z_r}{z_b - z_a} \left( 1 - \frac{z_b}{z_f} \right) & \frac{z_r}{z_b - z_a} \left( 1 - \frac{z_a}{z_f} \right) \\
                                                                 \frac{z_b}{z_b - z_a} & -\frac{z_a}{z_b - z_a} \end{bmatrix} \begin{bmatrix} x_a \\ x_b \end{bmatrix} \\
\end{align*} $$

---

Angle-dependent effect:
$$ \begin{align*}
    O\!\left( x_r, y_r, x_p, y_p \right) = \cos^4 \theta = \left( \frac{z_r^2}{\left( x_r - x_p \right)^2 + \left( y_r - y_p \right)^2 + z_r^2} \right)^2
\end{align*} $$

$$ \begin{align*}
    \varphi \text{ evaluations depends on } \begin{cases} \text{Display element}\\ \text{Focus distance}\\ \text{Position at retina}\end{cases}
\end{align*} $$

$$ \begin{align*}
    \text{Total number of rays before: }& \quad n_\text{focus} \times n_\text{elem} \times n_{rays}\\
    \text{Number of } \varphi \text{ evaluations per element before: }& \quad \phantom{n_\text{focus} \times n_\text{elem} \times .} n_{rays}\\
    \text{Total number of } \varphi \text{ evaluations before: }& \quad n_\text{focus} \times n_\text{elem} \times n_{rays}\\
    \\
    \text{Total number of rays now: }& \quad \phantom{n_\text{focus} \times n_\text{elem} \times .} N_{rays} = \quad \phantom{n_\text{focus} \times n_\text{elem} \times .} n_\text{elem} \times n_{rays}\\
    \text{Number of } \varphi \text{ evaluations per element now: }& \quad \phantom{n_\text{focus} \times n_\text{elem} \times .} N_{rays} = \quad \phantom{n_\text{focus} \times n_\text{elem} \times .} n_\text{elem} \times n_{rays}\\
    \text{Total number of } \varphi \text{ evaluations now: }& \quad n_\text{focus} \times n_\text{elem} \times N_{rays} = \quad n_\text{focus} \times n_\text{elem} \times \left( n_\text{elem} \times n_{rays} \right)\\
\end{align*} $$

```python
class Scene3D:
    def __init__(self, filename):
        self.trimesh = trimesh.load_mesh(filename)
        self.trimesh.apply_transform(trimesh.transformations.rotation_matrix(1.2, [0, 1, 0])) # <== Rotation
        if isinstance(self.trimesh, trimesh.Scene):
            self.meshes = list(self.trimesh.geometry.values())
        if isinstance(self.trimesh, trimesh.Trimesh):
            self.meshes = [self.trimesh]

        self.device = embree.Device()
        self.scene = self.device.make_scene()
        for mesh in self.meshes:
            geometry = self.device.make_geometry(embree.GeometryType.Triangle)
            vertex_buffer = geometry.set_new_buffer(embree.BufferType.Vertex, 0, embree.Format.Float3, 3 * np.dtype('float32').itemsize, mesh.vertices.shape[0])
            index_buffer = geometry.set_new_buffer(embree.BufferType.Index, 0, embree.Format.Uint3, 3 * np.dtype('uint32').itemsize, mesh.faces.shape[0])
            vertex_buffer[:] = mesh.vertices + [0.0, -35.0, 300.0] # <== Translation
            index_buffer[:] = mesh.faces
            geometry.commit()
            self.scene.attach_geometry(geometry)
            geometry.release()
        self.scene.commit()

    def rayCast(self, ray_origins, ray_directions):
        rayhit = embree.RayHit1M(ray_origins.shape[0])
        rayhit.org[:] = ray_origins
        rayhit.dir[:] = ray_directions
        rayhit.tnear[:] = 0
        rayhit.tfar[:] = np.inf
        rayhit.flags[:] = 0
        rayhit.prim_id[:] = embree.INVALID_GEOMETRY_ID
        rayhit.geom_id[:] = embree.INVALID_GEOMETRY_ID

        context = embree.IntersectContext()
        self.scene.intersect1M(context, rayhit)

        ray_colors = np.zeros((ray_origins.shape[0], 4), dtype=np.uint8)
        for mesh_id in range(len(self.meshes)):
            mask = rayhit.geom_id == mesh_id
            hit_uv = rayhit.uv[mask, np.newaxis, :]
            texture_uv = self.meshes[mesh_id].visual.uv[self.meshes[mesh_id].faces[rayhit.prim_id[mask], :], :]
            interpolated_uv = (1.0 - hit_uv[..., 0] - hit_uv[..., 1]) * texture_uv[:, 0, :] + hit_uv[..., 0] * texture_uv[:, 1, :] + hit_uv[..., 1] * texture_uv[:, 2, :]
            ray_colors[mask] = trimesh.visual.color.uv_to_color(interpolated_uv, self.meshes[mesh_id].visual.material.image)
        return ray_colors
    
    def release(self):
        self.scene.release()
        self.device.release()
```

```python
def circle_box_area(left, right, bottom, top):
    """Return the area of the intersection of a box against the unit circle."""
    # Clip horizontal intervals to [-1, 1].
    left = cp.clip(left, -1.0, 1.0)
    right = cp.clip(right, -1.0, 1.0)

    # Split vertical intervals into upper and lower halves.
    bottom_above = cp.clip(bottom, 0.0, 1.0)
    top_above = cp.clip(top, 0.0, 1.0)
    bottom_below = cp.clip(-top, 0.0, 1.0)
    top_below = cp.clip(-bottom, 0.0, 1.0)

    # Compute semicircle-box intersection areas.
    area_above = semicircle_box_area(left, right, bottom_above, top_above)
    area_below = semicircle_box_area(left, right, bottom_below, top_below)

    # Return the sum of the upper and lower halves.
    return area_above + area_below

def semicircle_box_area(left, right, bottom, top):
    """Return the area of the intersection of a box against the unit semicircle."""
    # Compute the positive x coordinate of a given y coordinate at the circle.
    bottom_intersection = semicircle(bottom)
    top_intersection = semicircle(top)

    # Compute lower and upper horizontal intervals.
    bottom_left = np.clip(left, -bottom_intersection, bottom_intersection)
    bottom_right = np.clip(right, -bottom_intersection, bottom_intersection)
    top_left = np.clip(left, -top_intersection, top_intersection)
    top_right = np.clip(right, -top_intersection, top_intersection)

    # Compute antiderivatives.
    bottom_left_antiderivative = semicircle_antiderivative(bottom_left)
    bottom_right_antiderivative = semicircle_antiderivative(bottom_right)
    top_left_antiderivative = semicircle_antiderivative(top_left)
    top_right_antiderivative = semicircle_antiderivative(top_right)

    # Compute areas.
    bottom_area = bottom_right_antiderivative - bottom_left_antiderivative
    top_area = top_right_antiderivative - top_left_antiderivative

    # Compute segment areas.
    bottom_segment = bottom_area - ((bottom_right - bottom_left) * bottom)
    top_segment = top_area - ((top_right - top_left) * top)

    # Compute area difference.
    return bottom_segment - top_segment

def semicircle(x):
    """Return the positive y (or x) coordinate associated to an x (or y) coordinate at the unit circle."""
    return cp.sqrt(1.0 - cp.square(x))

def semicircle_antiderivative(x):
    """Return the indefinite integral at x of the semicircle function."""
    return (cp.sqrt(1.0 - cp.square(x)) * x + cp.arcsin(x))/2
```

```python
def array_stats(a):
    print('shape: ', a.shape, a.dtype, '| total size:', a.nbytes/2**30, 'GiB | range: [', np.amin(a), ',', np.amax(a), '] | average: ', np.average(a))



def sampling_lattice(res, pitch=1.0):
    """Return an array of shape (res,) containing the coordinates of a sampling lattice with 0 as one of the samples."""
    return scipy_fft.fftshift(scipy_fft.fftfreq(res).astype(np.float32) * (res * pitch))



def display_image(values, color_map=None, absolute=True, normalize=True, shift=False, flip=False):
    """Display an array as an image using a color map. The image can be normalized, shifted and/or flipped."""
    values = values.T
    values = np.abs(values) if absolute else values
    values = scipy_fft.fftshift(values) if shift else values
    values = np.flip(values, axis=1) if flip else np.flip(values, axis=0)
    if len(values.shape) < 3:
        values = values * np.nan_to_num(1.0 / np.amax(values)) if normalize else np.clip(values, 0.0, 1.0)
        values = cm.get_cmap(color_map)(values, bytes=True)
    else:
        values = (values * np.nan_to_num(255.0 / np.amax(values))).astype(np.uint8) if normalize else np.clip(values * 255.0, 0.0, 255.0).astype(np.uint8)
    display(Image.fromarray(values))



def display_to_eye(z_r, x_a, z_a, x_b, z_b, z_f):
    """Transform coordinates from display space to eye space."""
    x_r = (x_a * (1.0 - z_b/z_f) - x_b * (1.0 - z_a/z_f)) * (-z_r) / (z_b - z_a)
    x_p = (x_a * z_b - x_b * z_a) / (z_b - z_a)

    return x_r, x_p



def eye_to_display(x_r, z_r, x_p, z_a, z_b, z_f):
    """Transform coordinates from eye space to display space."""
    x_a = x_r * z_a/z_r + x_p * (1.0 - z_a/z_f)
    x_b = x_r * z_b/z_r + x_p * (1.0 - z_b/z_f)

    return x_a, x_b



def evaluate_phi(x_r, y_r, z_r, x_a, y_a, z_a, w_a, x_b, y_b, z_b, w_b, z_f, size_p):
    # Ratios for projection of plane A and plane B to the pupil plane
    ratio_a = np.reciprocal(1.0 - z_a/z_f)
    ratio_b = np.reciprocal(1.0 - z_b/z_f)
    # Project display samples to the pupil plane
    x_a_at_p = (x_a - x_r * (z_a/z_r)) * ratio_a
    x_b_at_p = (x_b - x_r * (z_b/z_r)) * ratio_b
    y_a_at_p = (y_a - x_r * (z_a/z_r)) * ratio_a
    y_b_at_p = (y_b - x_r * (z_b/z_r)) * ratio_b
    # Lower and upper bounds.
    lower_x_a_at_p = x_a_at_p - np.abs(ratio_a * w_a/2)
    upper_x_a_at_p = x_a_at_p + np.abs(ratio_a * w_a/2)
    lower_x_b_at_p = x_b_at_p - np.abs(ratio_b * w_b/2)
    upper_x_b_at_p = x_b_at_p + np.abs(ratio_b * w_b/2)
    lower_y_a_at_p = y_a_at_p - np.abs(ratio_a * w_a/2)
    upper_y_a_at_p = y_a_at_p + np.abs(ratio_a * w_a/2)
    lower_y_b_at_p = y_b_at_p - np.abs(ratio_b * w_b/2)
    upper_y_b_at_p = y_b_at_p + np.abs(ratio_b * w_b/2)
    # Intersect the projections from A and B at the pupil.
    lower_x_p = cp.clip(lower_x_a_at_p, lower_x_b_at_p, upper_x_b_at_p)
    upper_x_p = cp.clip(upper_x_a_at_p, lower_x_b_at_p, upper_x_b_at_p)
    lower_y_p = cp.clip(lower_y_a_at_p, lower_y_b_at_p, upper_y_b_at_p)
    upper_y_p = cp.clip(upper_y_a_at_p, lower_y_b_at_p, upper_y_b_at_p)
    # Pupil radius.
    r = size_p/2
    # Return the normalized value of φ.
    return circle_box_area(lower_x_p/r, upper_x_p/r, lower_y_p/r, upper_y_p/r) / np.pi



def plot_display_elements(slopes, heights, intervals, size_r, size_p):
    x_r = np.linspace(-size_r/1.5, size_r/1.5, 1024).reshape(-1, 1, 1, 1) # shape = (n_r, n_elem, B/T, A/B)
    d1_phi = np.sum((x_r * slopes + heights) * (x_r > intervals[0]) * (x_r < intervals[1]) * [[-1], [1]], axis=(-2, -1))

    n_elem = heights.shape[0]

    slopes = np.tile(np.repeat(slopes, 3, axis=-1), 2)
    heights = np.repeat(heights.reshape(n_elem, -1), 3, axis=-1)
    intervals = np.stack(np.broadcast_arrays(intervals[0], intervals[1], np.nan), axis=-1).reshape(n_elem, -1)

    fig, axs = pyplot.subplots(2, constrained_layout=True, figsize=(12, 12))

    axs[0].set_xlabel(r'$x_r$', fontsize=14)
    axs[0].set_ylabel(r'$x_p$', fontsize=14)
    axs[0].set_aspect('equal')
    axs[0].set_xlim(-size_r/1.5, size_r/1.5)
    axs[0].set_ylim(-size_p/1.5, size_p/1.5)

    axs[0].plot([size_r/2, -size_r/2, -size_r/2, size_r/2, size_r/2], [size_p/2, size_p/2, -size_p/2, -size_p/2, size_p/2], 'k--')

    for id_elem in range(n_elem):
        axs[0].plot(intervals[id_elem], intervals[id_elem] * slopes + heights[id_elem])

    axs[1].set_xlabel(r'$x_r$', fontsize=14)
    axs[1].set_ylabel(r'$\varphi(x_r)$', fontsize=14)
    axs[1].set_aspect('equal')
    axs[1].set_xlim(-size_r/1.5, size_r/1.5)
    axs[1].set_ylim(0, 2*size_p/1.5)

    for id_elem in range(n_elem):
        axs[1].plot(np.squeeze(x_r), d1_phi[:, id_elem])

    axs[1].plot(np.squeeze(x_r), np.sum(d1_phi, axis=-1), 'black', zorder=10)
    axs[1].plot([size_r/2, -size_r/2, -size_r/2, size_r/2, size_r/2], [size_p, size_p, 0, 0, size_p], 'k--')



def vignetting(x_r, y_r, z_r, x_p, y_p):
    """Compute the vignetting effect, i.e. cos⁴ θ."""
    return np.square(np.reciprocal(np.square(x_r - x_p) + np.square(y_r - y_p) + np.square(z_r)) * np.square(z_r))
```

```python
%%time
print('Computing possible element overlapping at the retina...')

elem_overlap = cp.zeros((n_f, n_round_elem, n_round_elem), dtype=cp.bool8)

# Display coordinates
x_a = d1_a[id_round_x]
y_a = d1_a[id_round_y]
x_b = d1_b[id_round_x]
y_b = d1_b[id_round_y]

# For each focus distance...
for id_f in range(n_f):
    z_f = lattice_f[id_f]
    # φ size at retina.
    phi_size = (w_a * np.abs(1.0 - z_b/z_f) + w_b * np.abs(1.0 - z_a/z_f)) * (-z_r) / (z_b - z_a)
    # Retina coordinates.
    x_r = (x_a * (1.0 - z_b/z_f) - x_b * (1.0 - z_a/z_f)) * (-z_r) / (z_b - z_a)
    y_r = (y_a * (1.0 - z_b/z_f) - y_b * (1.0 - z_a/z_f)) * (-z_r) / (z_b - z_a)
    # Distance between elements over each axis separately.
    dist_x = cp.abs(x_r - x_r.reshape(-1, 1))
    dist_y = cp.abs(y_r - y_r.reshape(-1, 1))
    # Elements overlapping over each axis separately.
    overlap_x = dist_x < phi_size
    overlap_y = dist_y < phi_size
    # Elements overlapping over both axis simultaneously.
    elem_overlap[id_f] = cp.logical_and(overlap_x, overlap_y)



    if True:
        n_overlap = cp.sum(elem_overlap[id_f], axis=-1)
        max_overlap = cp.max(n_overlap)
        n_overlap = n_overlap / max_overlap

        fig, ax = pyplot.subplots(1, constrained_layout=True, figsize=(12, 12))

        ax.set_xlabel(r'$x_r$', fontsize=14)
        ax.set_ylabel(r'$y_r$', fontsize=14)
        ax.set_aspect('equal')
        ax.set_xlim(-size_r/1.3, size_r/1.3)
        ax.set_ylim(-size_r/1.3, size_r/1.3)

        box_h = cp.array([1, -1, -1,  1, 1]) / 2
        box_v = cp.array([1,  1, -1, -1, 1]) / 2

        ax.plot((box_h * size_r).get(), (box_v * size_r).get(), 'black', zorder=10)

        for id_elem in trange(n_round_elem):
            color = cm.get_cmap(None)(n_overlap[id_elem].get(), bytes=False)
            ax.plot((x_r[id_elem] + box_h * phi_size).get(), (y_r[id_elem] + box_v * phi_size).get(), color=color)

array_stats(elem_overlap)
array_stats(cp.sum(elem_overlap, axis=-1))
```

```python
%%time
print('Casting and processing rays...')

rng = cp.random.default_rng()

retina_colors = np.zeros((n_f, 3, n_r, n_r), dtype=np.float32)
retina_counts = np.zeros((n_f, 1, n_r, n_r), dtype=np.int32)

naive_colors = cp.zeros((3, n_round_elem), dtype=cp.float32)

elem_colors = cp.zeros((n_f, 3, n_round_elem), dtype=cp.float32)
elem_counts = cp.zeros((n_f, 1, n_round_elem), dtype=cp.int32)

scene = Scene3D('../models/teapot/teapot.obj')

# For each display element...
for id_elem in trange(n_round_elem):
    # Separable axis element IDs.
    id_x = id_round_x[id_elem]
    id_y = id_round_y[id_elem]
    # Sample rays uniformly inside element.
    ray_x_a = d1_a[id_x] + (rng.random(n_rays) - 0.5).astype(cp.float32) * w_a
    ray_x_b = d1_b[id_x] + (rng.random(n_rays) - 0.5).astype(cp.float32) * w_b
    ray_y_a = d1_a[id_y] + (rng.random(n_rays) - 0.5).astype(cp.float32) * w_a
    ray_y_b = d1_b[id_y] + (rng.random(n_rays) - 0.5).astype(cp.float32) * w_b
    # Ray origins and directions for ray cast.
    ray_origins = np.stack(np.broadcast_arrays(ray_x_a.get(), ray_y_a.get(), z_a), axis=-1)
    ray_directions = np.stack(np.broadcast_arrays(ray_x_b.get(), ray_y_b.get(), z_b), axis=-1) - ray_origins
    # Ray cast scene.
    ray_colors = cp.array(np.moveaxis(scene.rayCast(ray_origins, ray_directions)[..., 0:3] / 255.0, -1, 0).astype(np.float32))

    # Pupil intersection.
    ray_x_p = (z_b * ray_x_a - z_a * ray_x_b) / (z_b - z_a)
    ray_y_p = (z_b * ray_y_a - z_a * ray_y_b) / (z_b - z_a)
    # Pupil mask.
    pupil_mask = (cp.square(ray_x_p) + cp.square(ray_y_p)) < cp.square(size_p/2)

    # Display element naive coefficient.
    naive_colors[:, id_elem] =  cp.sum(ray_colors, axis=-1) / n_rays

    # For each focus distance...
    for id_f in range(n_f):
        z_f = lattice_f[id_f]
        # Retina intersection.
        ray_x_r = (ray_x_a * (1 - z_b/z_f) - ray_x_b * (1 - z_a/z_f)) * (-z_r) / (z_b - z_a)
        ray_y_r = (ray_y_a * (1 - z_b/z_f) - ray_y_b * (1 - z_a/z_f)) * (-z_r) / (z_b - z_a)
        # Retina mask.
        retina_mask = cp.logical_and(cp.abs(ray_x_r) < size_r/2, cp.abs(ray_y_r) < size_r/2)

        # Eye mask.
        eye_mask = 1.0 * cp.logical_and(pupil_mask, retina_mask)
        # Filtered rays.
        eye_ray_colors = ray_colors * eye_mask

        # Retina element id.
        id_x_r = cp.clip(cp.floor_divide(ray_x_r, w_r) + n_r/2, 0, n_r - 1).astype(cp.uint32).get()
        id_y_r = cp.clip(cp.floor_divide(ray_y_r, w_r) + n_r/2, 0, n_r - 1).astype(cp.uint32).get()
        # CPU copies.
        cpu_ray_colors = eye_ray_colors.get()
        cpu_eye_mask = eye_mask.get().astype(np.int32)
        # For each ray...
        for id_ray in range(n_rays):
            retina_colors[id_f, :, id_x_r[id_ray], id_y_r[id_ray]] += cpu_ray_colors[:, id_ray]
            retina_counts[id_f, :, id_x_r[id_ray], id_y_r[id_ray]] += cpu_eye_mask[id_ray]

        # Overlapping element IDs.
        id_overlap = cp.nonzero(elem_overlap[id_f, id_elem])[0]

        # Sample coordinates of overlapping elements.
        x_a = d1_a[id_round_x[id_overlap]].reshape(-1, 1)
        y_a = d1_a[id_round_y[id_overlap]].reshape(-1, 1)
        x_b = d1_b[id_round_x[id_overlap]].reshape(-1, 1)
        y_b = d1_b[id_round_y[id_overlap]].reshape(-1, 1)

        # Evaluate each overlapping element φ for every ray.
        phi_ray = evaluate_phi(ray_x_r, ray_y_r, z_r, x_a, y_a, z_a, w_a, x_b, y_b, z_b, w_b, z_f, size_p)

        # How many rays intersect each φ support.
        elem_counts[id_f][:, id_overlap] += cp.count_nonzero(phi_ray, axis=-1)
        # Compute ΦᵀR.
        elem_colors[id_f][:, id_overlap] += cp.sum(phi_ray * ray_colors.reshape(3, 1, n_rays), axis=-1)
        # Compute ΦᵀΦ.

scene.release()

array_stats(retina_counts)
array_stats(retina_colors)

array_stats(naive_colors)

array_stats(elem_counts)
array_stats(elem_colors)

# Divide the accumulated retina colors by the number of rays.
retina_colors = retina_colors / retina_counts

jureg

# Divide the accumulated element colors by the number of rays and multiply it by the support area of φ.
for id_f in range(n_f):
    zero_x, zero_y = cp.nonzero(elem_counts[id_f, 0] == 0)
    array_stats(elem_counts[id_f, :, zero_x, zero_y])
    array_stats(elem_colors[id_f, :, zero_x, zero_y])

    nonzero_x, nonzero_y = cp.nonzero(elem_counts[id_f, 0])
    array_stats(elem_counts[id_f, :, nonzero_x, nonzero_y])
    array_stats(elem_colors[id_f, :, nonzero_x, nonzero_y])

    elem_colors[id_f, :, nonzero_x, nonzero_y] = support_size[id_f, nonzero_x, np.newaxis] * support_size[id_f, nonzero_y, np.newaxis] * elem_colors[id_f, :, nonzero_x, nonzero_y] / elem_counts[id_f, :, nonzero_x, nonzero_y]

# Average values over the focus distance.
avg_colors = cp.average(elem_colors, axis=0)
```

```python
%%time
print('Computing non-negative display coefficients through the multiplicative update rule...')

opt_colors = 1.0 - rng.random((n_f, 3, n_d1_elem, n_d1_elem), dtype=cp.float64)

for id_f in range(n_f):
    for iteration in range(10):
        opt_colors[id_f] = opt_colors[id_f] * elem_colors[id_f] / (d1_phi_autocov[id_f] @ opt_colors[id_f] @ d1_phi_autocov[id_f].T + cp.finfo(cp.float64).eps)



opt_avg_colors = 1.0 - rng.random((3, n_d1_elem, n_d1_elem), dtype=cp.float64)

for iteration in range(10):
    opt_avg_colors = opt_avg_colors * avg_colors / (d1_avg_phi_autocov @ opt_avg_colors @ d1_avg_phi_autocov.T + cp.finfo(cp.float64).eps)



# For each focus distance...
for id_f, z_f in np.ndenumerate(lattice_f):
    print(' \n \nOptimal coefficients for focus distance =', z_f)
    array_stats(opt_colors[id_f])
    display_image(opt_colors[id_f].get(), normalize=True)

print(' \n \nOptimal coefficients for all focus distance')
array_stats(opt_avg_colors)
display_image(opt_avg_colors.get(), normalize=True)
```

```python
%%time
print('Computing display coefficients in [0, 1] through the clipping multiplicative update rule...')

clipped_opt_colors = 1.0 - rng.random((n_f, 3, n_d1_elem, n_d1_elem), dtype=cp.float64)

for id_f in range(n_f):
    for iteration in range(10):
        clipped_opt_colors[id_f] = cp.clip(clipped_opt_colors[id_f] * elem_colors[id_f] / (d1_phi_autocov[id_f] @ clipped_opt_colors[id_f] @ d1_phi_autocov[id_f].T + cp.finfo(cp.float64).eps), 0.0, 1.0)



clipped_opt_avg_colors = 1.0 - rng.random((3, n_d1_elem, n_d1_elem), dtype=cp.float64)

for iteration in range(10):
    clipped_opt_avg_colors = cp.clip(clipped_opt_avg_colors * avg_colors / (d1_avg_phi_autocov @ clipped_opt_avg_colors @ d1_avg_phi_autocov.T + cp.finfo(cp.float64).eps), 0.0, 1.0)



# For each focus distance...
for id_f, z_f in np.ndenumerate(lattice_f):
    print(' \n \nOptimal clipped coefficients for focus distance =', z_f)
    array_stats(clipped_opt_colors[id_f])
    display_image(clipped_opt_colors[id_f].get(), normalize=True)

print(' \n \nOptimal clipped coefficients for all focus distance')
array_stats(clipped_opt_avg_colors)
display_image(clipped_opt_avg_colors.get(), normalize=True)
```
