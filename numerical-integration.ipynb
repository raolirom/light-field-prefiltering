```python
import numpy as np
from matplotlib import pyplot



def f(x):
    x = np.abs(x)
    return (5/4) * np.power(1 - x, 3) * (3 * x + 1) * (x < 1.0)



def plot(x, samples, coordinates, partials, size):
    fig, ax = pyplot.subplots(1, constrained_layout=True, figsize=(12, 4))

    ax.plot(x, f(x), color='C0')
    ax.scatter(samples, f(samples), color='C1')

    ax.axvline(-size/2, color='C2')
    ax.axvline( size/2, color='C2')
    ax.axvline(-1.0, color='C3')
    ax.axvline( 1.0, color='C3')

    ax.hist(coordinates, bins=len(coordinates), range=(-size/2, size/2), weights=partials, fill=False, edgecolor='C4')

    pyplot.show()



size = 6.0
resolution = 25
pitch = size/resolution

x = np.linspace(-size/2, size/2, num=1000)
coordinates = (np.arange(resolution) - (resolution - 1)/2) * pitch

full_error_sum = 0.0
support_error_sum = 0.0
partial_error_sum = 0.0

rng = np.random.default_rng()

iterations = 50
for k in range(iterations):
    samples = rng.random(100) * size - size/2
    values = f(samples)

    support_incidence = np.logical_and(samples > -1, samples <= 1)

    partial_incidence = np.logical_and(samples > (coordinates.reshape(-1, 1) - pitch/2), samples <= (coordinates.reshape(-1, 1) + pitch/2))
    partials = np.sum(values * partial_incidence, axis=1) / np.maximum(np.sum(partial_incidence, axis=1), 1.0)

    plot(x, samples, coordinates, partials, size)

    full_result = np.average(values) * size
    support_result = 2 * np.sum(values * support_incidence) / np.sum(support_incidence)
    partial_result = np.sum(partials) * pitch

    full_error = abs(1.0 - full_result)
    support_error = abs(1.0 - support_result)
    partial_error = abs(1.0 - partial_result)

    full_error_sum += full_error
    support_error_sum += support_error
    partial_error_sum += partial_error

    print('   Full range solution = {} \t Error = {}'.format(full_result, full_error))
    print('Support range solution = {} \t Error = {}'.format(support_result, support_error))
    print('      Partial solution = {} \t Error = {}'.format(partial_result, partial_error))
    print('\n\n')

print('   Full range solution average error =', full_error_sum/iterations)
print('Support range solution average error =', support_error_sum/iterations)
print('      Partial solution average error =', partial_error_sum/iterations)
```
