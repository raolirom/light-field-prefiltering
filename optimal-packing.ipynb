```python
from PIL import Image
import numpy as np
from matplotlib import pyplot, cm
from IPython.display import display

def compute_eta_b(z_f):
    return (1 - z_a/z_f) * z_r * xi_r / (z_b - z_a)

def compute_eta_a(z_f):
    return (1 - z_b/z_f) * z_r * xi_r / (z_a - z_b)

z_r = 17.0
z_a = 6.5
z_b = 15.5

z_near = 1.3*z_b
z_far = 4*z_b

xi_r = 1.0

eta_a_near = compute_eta_a(z_near)
eta_b_near = compute_eta_b(z_near)
eta_a_far = compute_eta_a(z_far)
eta_b_far = compute_eta_b(z_far)
```

```python
c = (z_near - z_a) * eta_a_far + (z_near - z_b) * eta_b_far
d = z_near * (2 * z_b - z_a) - z_a * z_b

xi_a = ((z_near - z_b) * z_r * xi_r - z_b * c) / d
xi_b = ((z_near - z_a) * z_r * xi_r - 0.5*z_a * c) / d

fig, ax = pyplot.subplots(figsize=(16, 16))

for x in np.arange(8) * 2 * xi_a:
    for y in np.arange(8) * 2 * xi_b:
        ax.fill([x-eta_a_far, x-eta_a_near, x+eta_a_near, x+eta_a_far, x-eta_a_far],
                [y-eta_b_far, y-eta_b_near, y+eta_b_near, y+eta_b_far, y-eta_b_far])

ax.set_aspect('equal')
```
