```python
import numpy as np
from matplotlib import pyplot

fig, axs = pyplot.subplots(10, 1, sharex=True, sharey=True, figsize=(20, 50))
fig.subplots_adjust(hspace=0.5)

x = np.linspace(-3.5, 3.5, 7001)
x_ = np.linspace(-0.5, 0.5, 1001)
sinc_ = np.sinc(x_)
unsinc_ = np.reciprocal(sinc_)
unsinc = np.append(np.tile(unsinc_[:-1], 7), unsinc_[-1])
sinc = np.sinc(x)
box = (np.abs(x) < 0.5) * 1.0

for i in range(len(axs)):
    axs[i].set_title("Figure {}".format(i+1), fontsize=14)
    axs[i].set_xlim(-3, 3)
    axs[i].set_yticks([0.0, 1.0])
    axs[i].set_xticks(np.linspace(-2.5, 2.5, 6))
    axs[i].xaxis.set_tick_params(labelbottom=True)
    axs[i].grid(True)

axs[0].add_patch(pyplot.Rectangle((-0.5, -0.05), 1.0, 0.1, linewidth=3, edgecolor='#d62728', facecolor='none', hatch='//'))

for i in np.linspace(-3, 3, 7):
    axs[1].add_patch(pyplot.Rectangle((i-0.5, -0.05), 1.0, 0.1, linewidth=3, edgecolor='#d62728', facecolor='none', hatch='//'))

axs[2].plot(x, box)
for i in np.linspace(-3, 3, 7):
    axs[2].add_patch(pyplot.Rectangle((i-0.5, -0.05), 1.0, 0.1, linewidth=3, edgecolor='#d62728', facecolor='none', hatch='//'))

axs[3].plot(x, sinc)
for i in np.linspace(-3, 3, 7):
    axs[3].add_patch(pyplot.Rectangle((i-0.5, -0.05), 1.0, 0.1, linewidth=3, edgecolor='#d62728', facecolor='none', hatch='//'))

axs[4].plot(x, sinc)
axs[4].plot(x, box, color='tab:blue', linestyle='dashed')
axs[4].fill_between(x, sinc, 1.0, where=np.abs(x) <= 0.5, color='pink')
axs[4].fill_between(x, sinc, where=np.abs(x) >= 0.5, color='orange')
for i in np.linspace(-3, 3, 7):
    axs[4].add_patch(pyplot.Rectangle((i-0.5, -0.05), 1.0, 0.1, linewidth=3, edgecolor='#d62728', facecolor='none', hatch='//'))

axs[5].plot(x_, unsinc_)
axs[5].plot(x_, sinc_, color='tab:blue', linestyle='dashed')
axs[5].add_patch(pyplot.Rectangle((-0.5, -0.05), 1.0, 0.1, linewidth=3, edgecolor='#d62728', facecolor='none', hatch='//'))

axs[6].plot(x, unsinc)
for i in np.linspace(-3, 3, 7):
    axs[6].add_patch(pyplot.Rectangle((i-0.5, -0.05), 1.0, 0.1, linewidth=3, edgecolor='#d62728', facecolor='none', hatch='//'))

axs[7].plot(x, sinc*unsinc)
for i in np.linspace(-3, 3, 7):
    axs[7].add_patch(pyplot.Rectangle((i-0.5, -0.05), 1.0, 0.1, linewidth=3, edgecolor='#d62728', facecolor='none', hatch='//'))

axs[8].plot(x, sinc*unsinc)
axs[8].plot(x, box, color='tab:blue', linestyle='dashed')
axs[8].fill_between(x, sinc*unsinc, where=np.abs(x) >= 0.5, color='orange')
for i in np.linspace(-3, 3, 7):
    axs[8].add_patch(pyplot.Rectangle((i-0.5, -0.05), 1.0, 0.1, linewidth=3, edgecolor='#d62728', facecolor='none', hatch='//'))

axs[9].plot(x, sinc*unsinc)
axs[9].plot(x, sinc, color='tab:blue', linestyle='dashed')
axs[9].fill_between(x, sinc*unsinc, sinc, where=np.abs(x) >= 0.5, color='tab:olive')
for i in np.linspace(-3, 3, 7):
    axs[9].add_patch(pyplot.Rectangle((i-0.5, -0.05), 1.0, 0.1, linewidth=3, edgecolor='#d62728', facecolor='none', hatch='//'))
```
