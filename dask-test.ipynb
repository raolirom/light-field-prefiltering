```python
import cupy as cp
import dask.array as da
```

```python
cpu_primal = da.random.random((16000, 16000))
cpu_primal
```

```python
cpu_dual = da.fft.fft(da.fft.fft(cpu_primal.rechunk((-1, 'auto')), axis=0).rechunk(('auto', -1)), axis=1)
cpu_dual
```

```python
gpu_primal = cpu_primal.map_blocks(cp.asarray)
gpu_primal
```

```python
gpu_dual = da.fft.fft(da.fft.fft(gpu_primal.rechunk((-1, 'auto')), axis=0).rechunk(('auto', -1)), axis=1)
gpu_dual
```

```python
%timeit cpu_dual.compute()
```

```python
%timeit gpu_dual.compute()
```
